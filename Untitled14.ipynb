```python
step  = 200
first = 0
cl = ['rs', 'cs', 'ms', 'gs', 'ys', 'ks', 'bs']
fig = plt.figure(figsize=(20, 12))
ax = plt.subplot(111)
for i in range (7):
    ax.plot(a[first + i * step : first  + i * step + step, 0], a[first + i * step: first + i * step + step, 1], cl[i])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("время поиска 5 ближайших соседей от числа признаков")
ax.legend(cl)
ax.grid()
```
